## Key insights

- Python + Pandas allow efficient analysis of large datasets.
- The goal of analysis is not just calculating numbers, but understanding the patterns behind them.
- In this Titanic example, we explored survival rates by gender and passenger class.
- Women and higher-class passengers had significantly higher survival probabilities.

## Ключевые выводы

- Python + Pandas позволяют эффективно анализировать большие наборы данных.

- Цель анализа — не просто вычисление чисел, а понимание закономерностей, лежащих в их основе.

- В этом примере с «Титаником» мы исследовали показатели выживаемости в зависимости от пола и класса пассажиров.

- У женщин и пассажиров более высокого класса вероятность выживания была значительно выше.
